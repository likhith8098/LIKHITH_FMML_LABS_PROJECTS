# SVM (Support Vector Machine)

**What is SVM ?**

Support Vector Machine'(SVM) is a supervised machine learning algorithm which can be used for both classification and regression. However, it is mostly used in classification problems.


In this algorithm, we plot each data item as a point in n-dimensional space (where n is the number of features) with the value of each feature being the value of a particular coordinate.

**What is Hyperplane ?**

  A hyperplane is defined as an n-1 dimensional Euclidean space that seperates an n-dimensional Euclidean space into two disconnected parts or classes. The "optimal" hyperplane is one that maximizes the margin between the two classes, which means the distance of the hyperplane is equal from both the nearest data points  of these two classes.





  In a two dimensional space, a hyperplane is a line that optimally divides the data points into two different classes. In three dimensions, a hyperplane is a 2D plane.

If data is perfectly seperable, there are infinite hyper plane

This is exactly how SVM works, it draws a decision boundary which is a hyperplane between any two classes in order to classify them.




The basic principle behind SVM is to draw a hyperplane that best seperates the two classes, so you start off by drawing a random hyperplane and then you check the distance between the hyperplane and the closest data points from each class.

**These classifiers  depend on support vectors only**

Normally the hyperplane which has the maximum distance from the support vectors is the most optimum hyperplane and this distance between the hyperplane and the support vectors is known as the margin.



**Limitations**



*  Maximal margin classifier cannot be used if the two classes are not seperableby a hyperplane
